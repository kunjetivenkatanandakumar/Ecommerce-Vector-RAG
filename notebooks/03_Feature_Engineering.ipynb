```python
import pandas as pd
import numpy as np
import re
```

```python
df = pd.read_csv(r"C:\\Users\\kunja\\Downloads\\Ecommerce_Vector_RAG\\data\\processed\\amazon_products_clean.csv")

df.head()
```

```python
df.info()
```

```python
df["discount_percentage"] = (
    df["discount_percentage"]
      .astype(str)
      .str.replace("%", "", regex=False)
)

df["discount_percentage"] = pd.to_numeric(
    df["discount_percentage"],
    errors="coerce"
)
```

```python
df["discount_amount"] = (
    df["actual_price"]
    - df["discounted_price"]
)
```

```python
df["description_length"] = (
    df["about_product"]
    .fillna("")
    .str.len()
)
```

```python
df["review_length"] = (
    df["review_content"]
    .fillna("")
    .str.len()
)
```

```python
df["product_name_length"] = (
    df["product_name"]
    .str.len()
)
```

```python
df["category_depth"] = (
    df["category"]
      .astype(str)
      .str.count(r"\|")
      + 1
)
```

```python
df["combined_text"] = (
    "Product Name: " + df["product_name"] +
    ". Category: " + df["category"] +
    ". Description: " + df["about_product"] +
    ". Review Title: " + df["review_title"] +
    ". Review: " + df["review_content"] +
    ". Rating: " + df["rating"].astype(str)
)
```

```python
df["combined_text"] = (
    df["combined_text"]
      .str.replace(r"\s+", " ", regex=True)
      .str.strip()
)
```
